### 6-2. 반복문
- for, while

#### while 
- 어떤 조건이 만족할 때까지 블록 수행
- 해당 조건이 거짓이면 수행을 멈춘다
    - 언제 빠져나오는지 정확히 알아야 한다.
- for + if

#### for vs. while
- for loop: 반복 횟수 정확히 알고, 반복 횟수 변하지 않을 때
    - 예: 전체 학생 성적 산출
- while loop: 반복 횟수 정확하지 않고, 어떤 조건을 만족시킬 때 프로그램을 종료하고 싶다면 사용
    - 예: 가위바위보로 이기면 종료

```python
for i in range(10):
    print(i)
```

```python
j = 0
while j < 5: # 0, 1, 2, 3, 4
    print(j) # 0, 1, 2, 3, 4
    j += 1   # 0, 1, 2, 3, 4, 5
```

```python
# 지금의 j값은?
j
```

```python
characters = ['a', 'b', 'c']
for ch in characters:
    print(ch)
```

```python
ch = 0
while ch < len(characters):
    print(characters[ch])
    ch += 1
```

### 6-3. 반복문의 제어
- 의도치 않게 코드를 종료시킬 수 있기 때문에 되도록 사용하지 않는다. 

- break: 논리적으로 반복을 종료시키는 방법
- else: 어떤 조건이 "완전히" 끝났을 때 한번 더 실행하는 블록. break로 끝난 것과 구분할 수 있다. 
- continue: 특정 조건을 건너뛰고 다음 조건을 수행함

#### * break

```python
for i in range(10):
    if i == 16: # 6일 때는"ended" 출력 안됨
        break
    print(i)
else:
    print("ended")
```

#### * continue

```python
for i in range(10):
    if i == 6: # 6만 건너뜀
        continue
    print(i)
else:
    print("ended")
```

#### * else

```python
j = 0 
while j < 5: # 0, 1, 2, 3, 4
    print(j) # 0, 1, 2, 3, 4
    if j == 4: break
    j += 1   # 1, 2, 3, 4, 5
else:
    print("")
```

### <구구단 예제>
- 1. for, while 선택한 뒤, n단을 입력받아서 n단을 출력한다.
    - 출력: === 3단 ===
            3 x 1 = 3
            3 x 2 = 6
            ...
            3 x 9 = 27
            
- 2. 1 ~ n단까지 출력한다.
- 3. 시작단, 끝단 둘다 입력받을 것
    - 예: 4 ~ 7단
    - 시작단이 끝단보다 작을 때 어떻게 처리할지
- 4. 프로그램으로 만든다. 
    - 1) n단  2) 1-n단  3) n-m단
    - 3번 -> n, m -> ...
    - 메뉴에 없는 선택 처리

```python
while True:
    print("1) n단 출력하기 \n2) 1 ~ n단 출력하기 \n3) n ~ m단 출력하기 \n4) 종료하기")
    menu = int(input("메뉴 번호를 입력하세요: "))
    
    if menu == 1:
        start_dan = int(input('단을 입력하세요: '))
        end_dan = start_dan
    elif menu == 2:
        start_dan = 1
        end_dan = int(input("끝단을 입력하세요: "))
    elif menu == 3:
        start_dan = int(input("시작단을 입력하세요: "))
        end_dan = int(input("끝단을 입력하세요: "))
    elif menu == 4:
        print("구구단을 종료합니다.")
        break
    else: 
        print("메뉴에 없는 번호를 입력하셨습니다. 다시 입력해주세요.")
        continue
    
    if start_dan < 1:
        print("시작단이 1보다 작습니다. 다시 입력해주세요.")
        continue
    if start_dan > end_dan:
        print("시작단이 끝단보다 작습니다. 다시 입력해주세요.")
        continue
    
    if end_dan > 9:
        print("끝단이 9보다 큽니다. 다시 입력해주세요.")
        continue
        
    print()
    for i in range(start_dan, end_dan+1):
        print(f"=== {i}단 ===")
        for j in range(1, 10):
            print(f'{i} * {j} = {i*j}')
        print()
    
    print('-----------------------\n')
```

```python
print("=== 3단 ===")
for i in range(1, 10):
    print(f"3 x {i} = {3*i}")
```

#### Quiz. 구텐버그의 각 파일을 돌아가면서 글자 수, 단어 수, 문장 수를 출력합니다.

```python
"""
austen-emma | 글자 수 = 12개, 단어 수 = 14개, 문장 수 = 15개 
...
"""
for fileid in gutenberg.fileids():
    num_raw = len(gutenberg.raw(fileid))
    num_words = len(gutenberg.words(fileid))
    num_sents = len(gutenberg.sents(fileid))
    
    print(f"{fileid[:-4]} | 글자 수 = {num_raw}개, 단어 수 = {num_words}, 문장 수 = {num_sents}개")
```

### 6-6. Brown
- categories (file id 대신)

```python
# from nltk.corpus import brown 
nltk.download('brown')
brown_corpus = nltk.corpus.brown
```

```python
brown_corpus.categories()
```

```python
# 괄호 안에 fileid도 작성할 수 있기 때문에, 변수명을 꼭 지정해야 함
humor = brown_corpus.words(categories='humor') 
```

```python
# modal verbs의 쓰임
modals = ['can', 'could', 'may', 'might', 'will', 'would']

fdist_brown = nltk.FreqDist([w.lower() for w in humor])

for verb in modals:
    print(verb, fdist_brown[verb])
```

### 6-7. conditional frequency distribution (CFD)
- 조건이 붙은 fdist

```python
cfd = nltk.ConditionalFreqDist(
# genre에 따른 fdist
    (genre, word.lower())
    for genre in brown_corpus.categories()
    for word in brown_corpus.words(categories=genre)
)
```

```python
cfd['adventure']['the']
```

```python
cfd['adventure'].most_common(10)
```

```python
# 교수님 이 부분에서 커널 실행이 멈추지 않습니다.. 어디서 오류가 나는 건지 잘 모르겠습니다.
cfd.tabulate(conditions=brown_corpus.categories()[:3], smaples=modals)
```
