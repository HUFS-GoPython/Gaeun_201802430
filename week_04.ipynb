### 5-4 NLTK
- Frequency Distribution: 단어가 몇 번 출현했는지 (빈도수)
- 텍스트의 성질 알기 위해
    - 단어 개수
    - 단어 빈도수

```python
import nltk
from nltk.book import *
nltk.download('book', quiet=True)
```

```python
fdist = FreqDist(text1) # cased
```

```python
fdist.most_common(10) # 제일 빈번한거 상위 10개
```

```python
fdist['that'], text1.count('that') # 두개가 같은 결과
```

```python
fdist.freq('that')
```

```python
fdist.plot(50, cumulative=True)
```

```python
# 가장 빈번하지 않은 => 1번 나타나는 단어들
    # 알파벳만 리스트에 담아보기, 중복없이
[word for word in fdist.hapaxes() if word.isalpha()]
```

```python
# 대소문자 구분 없이 freq dist 만드려면? the 15,000 The 600 => the 21,000
    # 리스트 -> 리스트
    # set -> 중복 없앰 -> 빈도수 == 1
fdist2 = FreqDist([word.lower() for word in text1])
fdist2['the']
```

```python
fdist2['The']
```

```python
fdist = FreqDist(text1) # 대소문자, 첫글자 대문자 the, The, THE
fdist2 = FreqDist([word.lower() for word in text1]) # 한꺼번에 세어질 수 있도록
```

```python
# text2에서 길이가 3자이하이고, 빈도수가 5이상
[word for (word, count) in FreqDist(text2).items()\
    if (len(word) <= 3) and (count >= 5)] 
    # dictionary형은 items() 뒤에 붙여줘야 함
```

### 5-5 리스트 <-> 스트링
- split(): string.split(separator)
    - default = ' '
- join(): separator.join(string)

```python
sent = 'I, have, a, car'
# ['I', 'have', 'a', 'car'] # tokens
sent.split(',') # string.strip() 이므로 erro
```

```python
# solution
sent_tokens = [word.strip() for word in sent.split(',')]
sent_tokens
```

```python
# ['I', 'have', 'a', 'car'] => 'I-have-a-car'
'-'.join(sent_tokens)
```
